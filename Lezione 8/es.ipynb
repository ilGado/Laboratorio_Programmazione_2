## Es 1

### Hai a disposizione un file data.csv contenente dati mensili di traffico aereo con due colonne:

### - date: data in formato YYYY-MM (mese/anno)
### - passengers: numero di passeggeri per quel mese
### Costruisci un modello di regressione polinomiale che approssima l’andamento del numero di passeggeri nel tempo.

### Carica il dataset.
### 1. Convertilo in un formato numerico utilizzando una colonna mese_numerico che conti i mesi a partire da gennaio 1949.
### 2. Applica una regressione polinomiale (grado a tua scelta).
### 3. Calcola l’RMSE tra i valori reali e quelli predetti.
### 4. Visualizza i dati reali e la curva stimata con Plotly.

```python
import pandas as pd
import numpy as np


df = pd.read_csv("https://raw.githubusercontent.com/lauranenzi/LabProgrammazione_II_2026/main/data/data.csv")

df = df.dropna(subset=["passengers"])

df.head()
```

```python
# conversione in formato numerico con mese_numerico (conta i mesi a partire da gennaio 1949)

df["mese_numerico"] = range(len(df))
df["mese_numerico"]
```

```python
# regressione polinomiale con grado a scelta

from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

x = df["mese_numerico"].values.reshape(-1, 1)
y = df["passengers"].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=False)

model = make_pipeline(PolynomialFeatures(3), LinearRegression())
model.fit(x_train, y_train)
pred = model.predict(x_test)


```

```python
# RMSE tra valori reali e quelli predetti

from sklearn.metrics import root_mean_squared_error

rmse = root_mean_squared_error(y_test, pred)

print(f"RMSE: {rmse:.2f}")

```

```python
# grafico di dati reali e curva stimata

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=x_train.reshape(-1),
    y=y_train,
    name="dati reali",
    mode='lines+markers',
    line=dict(color="blue", width=4)
))

fig.add_trace(go.Scatter(
    x=x_test.reshape(-1),
    y=pred,
    name="curva stimata",
    mode='lines+markers',
    line=dict(color="red", width=4)
))

fig.update_layout(
    title="Grafico dei dati e della regressione lineare",
    xaxis_title="data",
    yaxis_title="passeggeri",
)

fig.show()
```
