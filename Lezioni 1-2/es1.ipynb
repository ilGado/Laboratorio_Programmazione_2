## 1. Trasformare cicli in list comprehension

### Fai tre esempi di cicli producono una lista e poi riscrivi lo stesso codice usando la list comprehension

```python
import numpy as np

lista1 = []

for i in range(100):
    lista1.append(i)
    
lista2 = np.arange(0, 100)
lista3 = np.linspace(0, 1, 11)
    
lista4 = [x for x in range(100)]

print(f"1: {lista1}\n2: {lista2}\n3: {lista3}\n4: {lista4}\n")
```

## 3 Operazioni su array

### Crea un array a, 1D (senza digitarlo esplicitamente), e genera due nuovi array b, contenente una sottostringa a piacere e c che è il reverse dell'array a (esempio [1,2,3] diventa [3,2,1]) . Dividi l'array a per l'array c. Fai la stessa cosa per una lista.

```python
a = np.random.randint(1, 100, 15)

print(f"a: {a}")
import random as rd

first = rd.randint(0, 14)
last = rd.randint(first, 14)

b = a[first:last]

print(f"b: {b}")

c = a[::-1]

print(f"c: {c}")

print(f"a / c = {a / c}")


lista = [rd.randint(1, 99) for i in range(15)]

print(f"lista: {lista}")

sottolista = lista[first:last]
print(f"sottolista: {sottolista}")

inversa = lista[::-1]

print(f"inversa: {inversa}")

res = [lista[i] / inversa[i] for i in range(15)]

print(f"lista / inversa = {res}")

```
