## 3. Seleziona il numero più vicino a 0.5 in un array casuale

### Genera un array 10 x 3 con numeri casuali nell'intervallo [0,1]. Per ogni riga, seleziona il numero più vicino a 0.5. Per farlo usa il fancy indexing.

```python
m = np.random.randint(0, 11, (10,3)) / 10
print(f"Matrix:\n{m}")

def getNearest(arr):
    
    diff = np.abs(arr - 0.5)   
    return np.argmin(diff) 

col = np.apply_along_axis(getNearest, axis=1, arr=m)
row = np.arange(10)
print(f"\nElementi più vicini a 0.5:\n{m[row,col]}")
```

## 4. Analisi della Frequenza Cardiaca con NumPy


### 4.1 Supponiamo che i seguenti valori rappresentino la tua frequenza cardiaca registrata dal tuo Fitbit durante la giornata: 68, 65, 77, 110, 160, 161, 162, 161, 160, 161, 162, 163, 164, 163, 162, 100, 90, 97, 72, 60, 70
Inserisci questi valori in un array NumPy.



```python
freq = np.array([68, 65, 77, 110, 160, 161, 162, 161, 160, 161, 162, 163, 164, 163, 162, 100, 90, 97, 72, 60, 70])
freq
```

### 4.2 Trova la frequenza cardiaca minima registrata durante la giornata.
(Questo valore rappresenta approssimativamente la tua frequenza cardiaca a riposo, un comune indicatore di salute.)

```python
np.min(freq)
```

### 4.3 Trova la frequenza cardiaca massima registrata durante la giornata.
(Questo valore è un'indicazione dell'intensità dell'esercizio fisico.)

```python
np.max(freq)
```

### 4.4 Calcoliamo la percentuale di letture effettuate mentre stavi facendo esercizio fisico (frequenza cardiaca sopra a 120):
    - Crea un nuovo array che contiene True quando la frequenza cardiaca è superiore a 120, e False quando è inferiore o uguale a 120.
    
    - Utilizza una funzione di sintesi per calcolare la percentuale di osservazioni in cui la tua frequenza cardiaca era sopra 120.

```python
isFreq = freq > 120

def calcFreq(arr):
    return round(arr.sum()/arr.size*100, 2)

print(f"Freq. cardiaca > 120: {calcFreq(isFreq)}%")
        
```

## 7. Analisi di una matrice

### Data la matrice survey con età, reddito e anni di istruzione


```python
survey_matrix = np.array([ [25, 40000, 10], [32, 52000, 12], [40, 63000, 14], [29, 47000, 11], [35, 58000, 13] ])
survey_matrix
```

### 7.1 Seleziona dalla matrice solo gli intervistati con almeno 12 anni di istruzione utilizzando un test logico.

```python
condiz = survey_matrix[:, 2] >= 12
survey_matrix[condiz]
```

### 7.2 In un'unica riga di codice, seleziona i redditi degli intervistati con almeno 12 anni di istruzione.

```python
reddito = survey_matrix[condiz][:,1]
reddito
```

### 7.3 In un'unica riga di codice, calcola il reddito medio degli intervistati con almeno 12 anni di istruzione.

```python
mediaReddito = round(np.mean(reddito), 4)
mediaReddito
```
